### **Ejercicio 2 (Bubble sort)**



#### Corrida de escritorio
$v_1=[3, 2, 5, 8, 4, 1]$

| i | Vector |
| -- | -- |
| 0 | $ [3, 2, 5, 8, 4, 1] $ |
| 1 | $ [3, 2, 5, 4, 1, 8] $|
| 2 | $ [3, 2, 4, 1, 5, 8] $|
| 3 | $ [3, 2, 1, 4, 5, 8] $|
...

Resultado final:

$v_{sorted} = [1,2,3,4,5,8]$

Casos de prueba:
* $v_2=[-1, 0, 4, 5, 6, 7]$
* $v_3$ 100_000 número aleatorios entre -200 y 145.

```python
def Ejercicio02(a):
    n = len(a)
    for i in range(n):
        swapped = False
        for j in range(1, n - i):
            if a[j] < a[j - 1]:
                a[j], a[j - 1] = a[j - 1], a[j]
                swapped = True
        if not swapped:
            break

# Ejemplo Clase
a = [3, 2, 5, 8, 4, 1]
Ejercicio02(a)
print(a)

print("Arreglo ordenado:", a)

```

* Modifique el algoritmo 02 y determine el número de comparaciones realizadas al ordenar la serie 5, 4, 3, 2, 1

```python
def Ejercicio02(a):
    n = len(a)
    comparaciones = 0  # contador 
    iteracion = 1      # mostrar iteración

    print(f"Arreglo inicial: {a}")

    for i in range(n):
        swapped = False
        print(f"\nIteración {iteracion}:")
        for j in range(1, n - i):
            # cada comparación entre elementos
            if a[j] < a[j - 1]:
                # si hay intercambio
                a[j], a[j - 1] = a[j - 1], a[j]
                comparaciones += 1
                swapped = True
                print(f"  Intercambio {comparaciones}: {a}")
        if not swapped:
            break
        iteracion += 1

    print(f"\nTotal de intercambios: {comparaciones}")
    return comparaciones
```

## Graficar!
* El valor de la serie $fib(n)$
* El valor del cociente 

    $\phi \rightarrow \frac{fib(n)} {fib(n-1)} \approx 1.618$ número áureo.


| n | $ fib(n) /fib(n-1) $ |
| -- | -- |
| 2 | $1/1=1 $ |
| 3 | $2/1 = 2$ |
| 4 | $3/2 = 1.5$ |
| 5 | $5/3= 1.66667$ |
| 6 | $8/5= 1.6$ |
| 7 | $13/8 = 1.625 $ |
| 8 | $21/13 = 1.615 $ |
| ... | ... |
|$\infty $ | $ \frac{1 + \sqrt{5}} {2} \approx 1.618$ (número áureo) 

```python
def Ejercicio03(n):
    if n == 0:
        return 0
    else:
        x = 0
        y = 1
        for i in range(1, n):
            z = x + y
            x = y
            y = z
        return y

n = 11
y = Ejercicio03(n)
print(f"Cuando n = {n}, y = {y}")
```

```python
import matplotlib.pyplot as plt

def Ejercicio03(n):
    if n == 0:
        return 0
    else:
        x = 0
        y = 1
        for i in range(1, n):
            z = x + y
            x = y
            y = z
        return y

n = 11

# Lista de valores fib(1), fib(2), ..., fib(n)
fib = [Ejercicio03(i) for i in range(1, n + 1)]

# Calculo de cocientes fib(n)/fib(n-1)
phi_aprox = [fib[i] / fib[i - 1] for i in range(1, len(fib))]

# Tabla
print("| n | fib(n)/fib(n-1) |")
print("|---|------------------|")
for i in range(1, len(fib)):
    print(f"| {i+1} | {fib[i]}/{fib[i-1]} = {phi_aprox[i-1]:.6f} |")

# Grafica Fibonacci
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(range(1, n+1), fib, marker='o')
plt.title("Serie de Fibonacci")
plt.xlabel("n")
plt.ylabel("fib(n)")
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(range(2, n+1), phi_aprox, marker='o', color='orange', label='fib(n)/fib(n-1)')
plt.axhline(y=(1 + 5 ** 0.5) / 2, color='red', linestyle='--', label='Número áureo φ')
plt.title("Aproximación al número áureo (φ)")
plt.xlabel("n")
plt.ylabel("Cociente")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

#### **Usando el Algoritmo 03 y la aritmética de redondeo con 3 cifras, determine la iteración desde la cual el error relativo de $\frac{y_{i+1}}{y_i} (i > 0)$ con respecto a $\frac{1 + \sqrt{5}}{2}$ está dentro de $10^{-5}$**

```python
import math

def round_sig(x, sig=3):
    """Redondea x a 'sig' cifras significativas."""
    if x == 0:
        return 0.0
    sign = 1 if x > 0 else -1
    x_abs = abs(x)
    exp = math.floor(math.log10(x_abs))
    factor = 10 ** (sig - 1 - exp)
    return sign * round(x_abs * factor) / factor


def Ejercicio03_round(sig=3, tol=1e-5, max_iter=50):
    """
    Calcula la sucesión de Fibonacci con redondeo a 'sig' cifras significativas.
    Determina a partir de qué iteración la razón fib(i+1)/fib(i)
    se aproxima al número áureo con error relativo < tol.
    """

    phi = (1 + math.sqrt(5)) / 2

    # valores iniciales redondeados
    x = round_sig(0.0, sig)
    y = round_sig(1.0, sig)

    print(f"{'i':>3} | {'fib(i)':>8} | {'fib(i+1)':>8} | {'razón':>10} | {'error relativo':>14}")
    print("-" * 55)

    found = None
    for i in range(1, max_iter + 1):
        # siguiente término redondeado
        z = round_sig(x + y, sig)

        if y != 0:
            ratio = z / y
            err_rel = abs(ratio - phi) / phi
        else:
            ratio, err_rel = float('inf'), float('inf')

        print(f"{i:3d} | {y:8.5g} | {z:8.5g} | {ratio:10.6f} | {err_rel:14.8e}")

        if err_rel < tol:
            found = i
            break

        x, y = y, z 

    if found:
        print(f"\n Con redondeo a {sig} cifras, el error relativo < {tol} a partir de la iteración {found}.")
    else:
        print(f"\n No se alcanzó el error < {tol} en {max_iter} iteraciones.")

    return found


# Ejecutar
Ejercicio03_round(sig=3, tol=1e-5, max_iter=50)
```

A qué valor converge el algoritmo 04?

```python
def Ejercicio04(n_max):
    
    suma = 0
    for n in range(1, n_max + 1):
        suma += 1/n
    return suma

# Suma para diferentes números de términos
n_valores = [10, 100, 1000, 10000, 100000, 1000000]

print("Resultado de la serie armónica Σ(1/n):")
for n in n_valores:
    resultado = Ejercicio04(n)
    print(f"Suma con {n} términos: {resultado:.6f}")


# El resultado de la serie armónica diverge, lo que significa que a medida que n aumenta, la suma se aproxima a infinito.
```

```python
import matplotlib.pyplot as plt

def Ejercicio04(n_max):
    suma = 0
    valores = []  # guardamos las sumas parciales
    for n in range(1, n_max + 1):
        suma += 1 / n
        valores.append(suma)
    return valores

# Valores de n a evaluar
n_valores = [10, 100, 1000]

print("Resultado de la serie armónica Σ(1/n):")
for n in n_valores:
    resultado = Ejercicio04(n)[-1]
    print(f"Suma con {n} términos: {resultado:.6f}")

# Graficar las sumas parciales
plt.figure(figsize=(9, 5))

for n in n_valores:
    y = Ejercicio04(n)
    x = range(1, n + 1)
    plt.plot(x, y, label=f'n = {n}', linewidth=1.5)  # línea
    plt.scatter(x, y, s=10)  # puntos

plt.title("Comportamiento de la Serie Armónica Σ(1/n)")
plt.xlabel("Número de términos (n)")
plt.ylabel("Suma parcial")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
